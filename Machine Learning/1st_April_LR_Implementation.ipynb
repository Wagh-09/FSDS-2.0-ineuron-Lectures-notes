```python
## Divide the dataset into independent and dependent 
x=df.iloc[:,:-1]   #all the rows and remove last column
y=df.iloc[:,-1]    #all the rows and only one column
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
x_train.head()
```

```python
x.shape
```

```python
x_train.shape,x_test.shape
```

```python
#Feature Scaling --Standardization for numerical features
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(x_train)        #fit function compute mean and sd from training data
```

```python
scaler.transform(x_train)        #find z-score
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_train
```

```python
#For x_test We want to find only transform function to compute 
```

```python
x_test=scaler.transform(x_test)
```

```python
#Model Training
#sklearn library has an open source library which has classification, Regression, Clusturing, dimension reduction,model selection,preprocessing
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
#Model fitting
regression.fit(x_train,y_train)       #To find coefficients
```

```python
#How many coefficients and slopes are there
regression.coef_
```

```python
regression.intercept_
```

```python
## prediction
y_pred=regression.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mse=mean_squared_error(y_test,y_test)
print(mse)
mae=mean_absolute_error(y_test,y_pred)
print(mae)
print(np.sqrt(mse))
```

```python
#To check the Accuracy of  model by R2 and adjusted R2
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_test,y_pred)
```
